```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sklearn import metrics
import math

```

```python
# storing X and Y into List/array
# X = input features, X1        # type = array/list          # size= 1x97
# y = output                    # type = array/list          # size= 1x97

file = open("ex1data1.txt","r")

X=[]     #create empty list
y=[]
for line in file:
  
  fields = line.split(",")
  
  X.append(float(fields[0]) )         #typecast str to float  
  y.append(float(fields[1]))          #append float to empty array

file.close() 

print(X)
print(y)
print(len(X))           #gives no of rows

new_list=list([[X],[y]])
print(new_list)

#print(len(new_list)) 
no_of_columns = len(new_list)
print(no_of_columns)
```

```python
plt.scatter(X,y,marker="x")
plt.xlabel("Population of city in 10000s ")
plt.ylabel("Profit in $10000s ")
```

```python
# x_0_row = 1x97  ; x_0_column = 97x1  ;  type= numpy array
# x_1_row = 1x97  ; x_1_column = 97x1  ;  type= numpy array
# input_features = 97x2  ;                type= numpy array
# y_row = 1x97    ;  y_column = 97x1   ;  type= numpy array

x_0_row = np.ones(len(X))
x_0_column = x_0_row.reshape(len(X),1)
print(x_0_row)
print(x_0_column)
print(len(x_0_row))
print(len(x_0_column))

x_1_row = np.array(X)
x_1_column = x_1_row.reshape(len(X),1)
print(x_1_row)
print(x_1_column)
print(len(x_1_row))
print(len(x_1_column))

input_features = np.concatenate((x_0_column, x_1_column), axis=1)
print(input_features)
print(len(input_features))

y_row = np.array(y)
y_column=y_row.reshape(len(y),1)
print(y_row)
print(y_column)
print(len(y_row))
print(len(y_column))
```

```python
# CostFunction
# theta_0 , theta_1 are constants (1x1)
# hypo_column : shape = 97x1 : type = numpy array
# hypo_row    : shape = 1x97 : type = numpy array
# error       : shape = 97x1 : type = numpy array

def CostFunction(theta_0,theta_1):

   hypo_column = np.dot(theta_0,x_0_column) + np.dot(theta_1,x_1_column)
   #print(hypo_column)

   hypo_row = hypo_column.T
   #print(hypo_row)

   error=np.subtract(hypo_column, y_column)
   #print(error)
 
   sq_error=np.square(error)
   #print(sq_error)

   sum=np.sum(sq_error,axis=0)
   #print(sum)

   m=len(error)
   #print(m)

   CF=sum/(2*m)
   #print(CF)
   
   return [CF,hypo_column]
```

```python
# Gradient Descent


def GradientDescent(theta_0,theta_1):

   alpha=0.01
  
   hypo_column = np.dot(theta_0,x_0_column) + np.dot(theta_1,x_1_column)
   #print(hypo_column)

   hypo_row = hypo_column.T
   #print(hypo_row)

   error=np.subtract(hypo_column, y_column)
   #print(error)
 
   m=len(error)
   #print(m)

   mul_0=np.multiply(error,x_0_column)
   mul_1=np.multiply(error,x_1_column)

   sum_0=np.sum(mul_0,axis=0)
   sum_1=np.sum(mul_1,axis=0)

   new_theta_0= theta_0 - ((alpha/m)*sum_0)
   new_theta_1= theta_1 - ((alpha/m)*sum_1)
   
   return new_theta_0,new_theta_1

```

```python
print(CostFunction(0,0)[0])
print(GradientDescent(0,0))
```

```python
new_theta_0=[]
new_theta_1=[]
newJ=[]

theta_0=0
theta_1=0
for i in range(0,1500):

  J,hypo = CostFunction(theta_0,theta_1)

  newJ.append(J)
  new_theta_0.append(theta_0)
  new_theta_1.append(theta_1)

  theta_0, theta_1 = GradientDescent(theta_0,theta_1)

  theta_0=float(theta_0)
  theta_1=float(theta_1)

  plt.plot(x_1_column,hypo)
  plt.xlabel("Population of city in 10000s ")
  plt.ylabel("Profit in $10000s ")
print(J)  
print(newJ)
print(new_theta_0)
print(new_theta_1)

```

```python
# theta0_vals : len = 100 ; type = numpy array ; shape = 100x1
# theta1_vals : len = 100 ; type = numpy array ; shape = 100x1
# J_vals : len =100       ; type = numpy array ; shape = 100x100

# Visualizing J(theta_0, theta_1):
# Grid over which we will calculate J

theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)

# initialize J_vals to a matrix of 0's
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

# Fill out J_vals
for i in range (0,len(theta0_vals)):
  for j in range (0,len(theta1_vals)):
    J_vals[i,j] = CostFunction(theta0_vals[i],theta1_vals[j])[0]

print(J_vals)
```
